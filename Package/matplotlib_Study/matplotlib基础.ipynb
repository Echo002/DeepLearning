```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure()
# 在图中画线
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')
plt.show()

plt.figure(num = 3, figsize = (8,5)) # 给参数定figure图的名称
plt.plot(x, y2)
plt.show()
```

```python
# 设置坐标轴

plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')

# 设置xy轴的取值范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 设置坐标轴名称
plt.xlabel('I am x')
plt.ylabel('I am y')

# 设置坐标轴刻度
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1, 0, 1, 2, 3],[r'$\alpha$',r'$b$','c','d','e','f'])
# 可以设置字体和转义字符

plt.show()
```

```python
# gca = 'get current axis'
ax = plt.gca()
plt.xticks(new_ticks)
plt.yticks([-2, -1, 0, 1, 2, 3],[r'$\alpha$',r'$b$','c','d','e','f'])

# 设置边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 设置默认坐标轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 移动坐标轴
ax.spines['bottom'].set_position(('data', -1))
ax.spines['left'].set_position(('data', -1))

plt.show()
```

```python
# 能见度
x = np.linspace(-3, 3, 50)
y = 0.1 * x

plt.figure()
plt.plot(x, y, linewidth = 10)
plt.ylim(-2, 2)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor = 'white', edgecolor = 'None', alpha = 0.7)) #alpha为透明度

plt.show()
```

```python
# 散点图
n = 1024
X = np.random.normal(0, 1, n) #平均值为0 方差为1
Y = np.random.normal(0, 1, n) #平均值为0 方差为1
T = np.arctan2(Y, X) #调颜色

plt.scatter(X, Y, s = 75, c = T, alpha = 0.5)
plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))

plt.xticks(())
plt.yticks(()) #去掉坐标轴的标注
plt.show()
```

```python
# 柱状图
n = 12
X = np.arange(n)
Y1 = (1 - X/float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X/float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, color = '#9999ff', edgecolor = 'white')
plt.bar(X, -Y2, color = '#ff9999', edgecolor = 'white')

for x, y in zip(X, Y1): #zip 打包成列表，元素为元组，元组的值为对应参数的值
    plt.text(x + 0.4, y + 0.05, '%.2f' %y, ha = 'center', va = 'bottom') # 前两个参数为位置 ha,va为横向和纵向对齐方式

for x, y in zip(X, Y2): #zip 打包成列表，元素为元组，元组的值为对应参数的值
    plt.text(x + 0.4, y - 0.05, '%.2f' %y, ha = 'center', va = 'up')
plt.show()
```
