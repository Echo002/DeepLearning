```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure()
# 在图中画线
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')
plt.show()

plt.figure(num = 3, figsize = (8,5)) # 给参数定figure图的名称
plt.plot(x, y2)
plt.show()
```

```python
# 设置坐标轴

plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')

# 设置xy轴的取值范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 设置坐标轴名称
plt.xlabel('I am x')
plt.ylabel('I am y')

# 设置坐标轴刻度
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1, 0, 1, 2, 3],[r'$\alpha$',r'$b$','c','d','e','f'])
# 可以设置字体和转义字符

plt.show()
```

```python
# gca = 'get current axis'
ax = plt.gca()
plt.xticks(new_ticks)
plt.yticks([-2, -1, 0, 1, 2, 3],[r'$\alpha$',r'$b$','c','d','e','f'])

# 设置边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 设置默认坐标轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 移动坐标轴
ax.spines['bottom'].set_position(('data', -1))
ax.spines['left'].set_position(('data', -1))

plt.show()
```
